## Investigation Overview
In this Analysis for Ford Gobike dataset, my main focus was on Time (months), Trip duration and User types and the relation between these factors.
## Dataset Overview
Ford GoBike System Dateset contains 2,506,983 bike rides info about individual rides made in a bike-sharing system covering the greater San Francisco Bay area through 2019. The dataset contains information about: 
- Trip duration: trip duration in seconds, start/end date/time for the bike rides. 
- Stations info: start/end station, name, geolocation (latitude/longitude) 
- Customer info: type of member (user_type)
- Bike id
- Bike Share for all Trip
- Rental access method

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df=pd.read_csv('2019_Tripdata.csv')
```

## How long does the average trip take?


```python
df["duration_min"] = round((df.duration_sec/60), 2)
bin_edges = np.arange(0, 36 + 3 , 3)
plt.hist(data = df, x = 'duration_min' , bins = bin_edges)
plt.title('Distribution of Trip Duration')
plt.xlabel('Trip Duration (Min)')
plt.ylabel('Counts');
```

## How long does the average trip take?

> Using log transformation

```python
ticks =  [ 1 , 3, 10, 30, 100, 300, 500]
labels = ['{}'.format(v) for v in ticks]
bins = 10**np.arange(0 , 3 + 0.1 , 0.1)
plt.hist(data = df, x = 'duration_min' , bins = bins);
plt.xscale('log');
plt.xticks(ticks,labels);
plt.xlabel('duration_min');
plt.title('Distribution of Trip Duration')
plt.ylabel('Counts');
```

## When are most trips taken in month of the year?

```python
sb.countplot(data=df, x='month', color=sb.color_palette()[0]);
plt.xlabel('Month of the trip')
plt.title('Number of trip in diffrent months');
```

## Bike Rides Percentage by user type

```python
df.user_type.value_counts().plot(kind='pie',autopct = '%1.1f%%',startangle = 90,
        counterclock = False,);
plt.title("Type of users", y=1.05, fontsize=14, fontweight='semibold')
```

## User type trips duration along the months

```python
plt.figure(figsize = [15, 5])
ax = sb.barplot(data = df, x = 'month', y = 'duration_min', hue = 'user_type')
ax.legend(loc = 8, ncol = 3, framealpha = 1, title = 'User Type');
plt.title('User type duration along the months');
```

## Summary
Ford Gobike dataset analysis output:
- The majority of trips durations around 10 mins and months with frequent rides are warm months such as July.
- Subscribers of the service seem to use the service of bike rides more than the Customers.
- subscribed user have higher number of trips than the customers specially during warm time.
- Customer users are using the bikes for longer period of time.
